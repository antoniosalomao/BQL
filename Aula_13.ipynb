```python
#-------#
# BqViz #
#-------#

import bql
import bqviz as bqv
import pandas as pd

bq = bql.Service()
params_dict = {    'dates': 'range(2016-01-01, 2017-01-01)',
                'currency': 'usd',
                    'fill':'next'}

# Data items: 
px_open = bq.data.px_open(**params_dict)
px_last = bq.data.px_last(**params_dict)
px_high = bq.data.px_high(**params_dict)
px_low = bq.data.px_low(**params_dict)

# Passing on 'column' dictionary: 
items_dict = { 'Open': px_open,
               'Last': px_last,
               'High': px_high,
                'Low': px_low}

# Universe:
bq_univ = 'IBM US Equity'

# Fetching data:
request = bql.Request(bq_univ, items_dict)
response = bq.execute(request)
df_main = bql.combined_df(response)

# Data wrangling:
df_main.set_index('DATE', inplace=True)
df_main.drop('CURRENCY', axis=1, inplace=True)
df_main.head()
```

```python
intlines = bqv.InteractiveLinePlot(df_main, hide_controls=True)
intlines.show()
```
