```python
import bql
import pandas as pd 
bq = bql.Service()

#-------------------#
#  Object oriented  #
#-------------------#

item_1 = bq.data.px_last(dates=bq.func.range('-5D', '0D'))
item_2 = bq.data.px_open(dates=bq.func.range('-5D', '0D'))
items = [item_1, item_2]
univ = ['FB US Equity', 'IBM US Equity']
request = bql.Request(univ, items)
response = bq.execute(request)
df = response[0].df()
df
```

```python
# Example 2:
date_range = bq.func.range('2020-01-01', '2020-08-17')
item_1 = bq.data.px_last(dates=date_range)
item_2 = bq.data.px_open(dates=date_range)
items = [item_1, item_2]
univ = ['FB US Equity', 'TSLA US Equity']
request = bql.Request(univ, items)
response = bq.execute(request)
df = bql.combined_df(response)
df.tail(10)
```

```python
# Data wrangling:
# PX_Last DataFrame

df.dropna(inplace=True, axis=0)
df = df.reset_index()
df.columns = ['ID', 'Date', 'Currency', 'PX_Last', 'PX_Open']
df_px_last = df.pivot('Date', 'ID', 'PX_Last')
df_px_open = df.pivot('Date', 'ID', 'PX_Open')
df = pd.concat([df_px_open, df_px_last], axis=1)
df.columns = ['FB US Equity - PX_Open',
              'TSLA US Equity - PX_Open',
              'FB US Equity	- PX_Close',
              'TSLA US Equity- PX_Close']
df.tail(10)
```
